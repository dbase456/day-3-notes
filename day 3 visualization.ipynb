```python
import pandas as pd
import seaborn
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(20)
y = np.random.normal(10,1,20)
z = np.random.normal(10,2,20)
print x
print y
z
```

```python
fig = plt.figure()
ax=fig.add_subplot(212)
## of rows, # of colmns, # chart #
ax.plot(x,y)
```

```python
#lines and scatter plots
```

```python
#crate two plots with line and scatterplot
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x,y)
ax2 = fig.add_subplot(122)
ax2.scatter(y,z)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.hist(y)
ax2.boxplot([y,z])
```

```python
#labels and legends

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y,label = 'Thing #1')
ax.plot(x,z,label = 'Thing #2')
ax.legend()
```

```python
#titles and suptitile

fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x,y)
ax1.set_title("Line")
ax2 = fig.add_subplot(122)
ax2.scatter(y,z)
ax2.set_title("Scatter")
fig.suptitle('deeps results')
plt.tight_layout()
```

```python
#axis labels and saving image
#Add x and y axis labels with font size

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x,y)
ax.set_ylabel('$$$Money',fontsize=10)
ax.set_xlabel('years',fontsize=20)

```

```python
fig.savefig('figure.png')
```
